```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 0.8*0.6*0.4       # transmission efficiency
Np = 10**4              # number of photon received from 0-magnitude star per s cm^2 nm
dlambda = 90            # filter bandpass (FWHM) from sensor(nm)
D = 200*10**(-1)        # telescope's effective diameter / effective aperture (mm) to (cm)
theta = 2               # seeing in observation site
QE = 0.8                # quantum efficiency
#Th = 0.81              # thermal noise
Th = 0                  # thermal noise ignored
R = 4                   # readout noise
eps = 0.37              # obscuration
```

```python
#pixel from object and background
npix = int(input("Number of pixel from object     : "))
nB   = int(input("Number of pixel from background : "))
```

```python
#sky background magnitude

mB = float(input("Magnitude of sky at observation location : "))
```

```python
#Range of estimated exposure time

initial_time = int(input("Initial exposure time : "))
ended_time   = int(input("Ended exposure time   : "))

#t_range = [i for i in range(initial_time,ended_time)]
t_range = np.linspace(initial_time,ended_time,1000*ended_time)
```

```python
#rate of photons from object
def F(V):
    F = Np*tau*(np.pi/4)*(D**2)*(1-(eps**2))*dlambda*(10**(-V/2.5))
    return F

#rate of photons from the background
def B(mB):
    B = Np*tau*(np.pi/4)*(D**2)*(1-(eps**2))*dlambda*(10**(-mB/2.5))*((np.pi/4)*(theta**2))
    return B

def eta(npix,nB):
    eta = npix*(1+(npix/nB))
    return eta

#Signal to Noise Ratio
def SNR(t,V,mB):
    SNR = (F(V)*QE*t)/(np.sqrt((F(V)*QE*t)+(B(mB)*QE*eta(npix,nB)*t)+(Th*eta(npix,nB)*t)+(eta(npix,nB)*(R**2))))
    return SNR
```

```python
xobject   = int(input("Number of object         : "))
N_object = []
V_object = []
for i in range(xobject):
    name  = str(input("Object name              : "))
    mag   = float(input("Object visual magnitude  : "))
    N_object.append(name)
    V_object.append(mag)
```

```python
List_SNR_i = []
for i in range(xobject):
    List_SNR = []
    for j in t_range:
        List_SNR.append(SNR(j, V_object[i], mB))
    List_SNR_i.append(List_SNR)
```

```python
color = ['orange','purple','olive','maroon','blue','red','green']
```

```python
for i in range(xobject):
    plt.plot(t_range, List_SNR_i[i], label=f"{N_object[i]}; V = {V_object[i]}", c = color[i])
plt.xlim(0,300)
plt.ylim(0,800)
plt.xlabel('exposure time (s)')
plt.ylabel('Signal to Noise Ratio')
plt.legend()
plt.grid()
plt.show()
```
